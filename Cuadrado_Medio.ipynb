```python
from tabulate import tabulate

print("")
print("Programa del Cuadrado Medio\n")

num_aleatorio = int(input("Escribe tu numero de 4 digitos"))
print("Tu numero aleatorio es: ",num_aleatorio)

#Guarda los resultados en una lista
resultado = []

for _ in range(100):
  numero_aleatorio = num_aleatorio * num_aleatorio

  cadena = str(numero_aleatorio)
  cantidad = len(cadena)
  cero = str(0)
  if cantidad < 8:
    numero_aleatorio2 = cero + cadena
    numero_aleatorio3 = [numero_aleatorio2[i] for i in range(1,5)]

    resultado.append(numero_aleatorio3)

  else:
    cadena_aleatoria =str(numero_aleatorio)
    numero_aleatorio = [cadena_aleatoria[i] for i in range(2,6)]

    resultado.append(numero_aleatorio)

  #Muestra resultado en una tabla
  headers = ["Iteracion", "Cuadrado Medio"]
  tabla_resultado = [[i + 1, ''.join(map(str, r)), int(''.join(map(str, r))) ] for i, r in enumerate(resultado)]

  print(tabulate(tabla_resultado, headers = headers, tablefmt = "fancy_grid"))
```
